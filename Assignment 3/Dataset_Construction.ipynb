```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
```

```python
# Data generation
input_tensor = torch.linspace(-10,10,1000)
sinc_values = torch.where(input_tensor !=0 , torch.sin(input_tensor)/input_tensor, torch.tensor(1.0))
sinc_values_after_noise = sinc_values + torch.randn(sinc_values.shape)

print(sinc_values_after_noise)
```

```python
# Visualization

plt.plot(input_tensor.numpy(),sinc_values.numpy(),color='blue',label='Sinc values before noise')
plt.scatter(input_tensor.numpy(),sinc_values_after_noise.numpy(),color='red',label='Sinc values after noise')
plt.xlabel('Input tensor')
plt.ylabel('Sinc vales(with / without noise)')
plt.title('Visualization')
plt.legend()
plt.show()
```
